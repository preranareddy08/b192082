```python
!pip install vaderSentiment

# Step 1: Import Libraries
import pandas as pd
import spacy
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


```

```python
# Step 2: Load Data
# Replace these paths with your actual file paths
media_data = pd.read_excel("/content/Media & Research Articles data.xlsx")
twitter_data = pd.read_excel("/content/Twitter Posts Data.xlsx")

```

```python
# Step 3: Explore and Clean Data
# Adding a 'Source' column to track data origin
media_data['Source'] = 'Media'
twitter_data['Source'] = 'Twitter'

```

```python
# Renaming columns for consistency
media_data = media_data.rename(columns={
    'Article title': 'Title',
    'Content': 'Body'
})
```

```python
twitter_data = twitter_data.rename(columns={
    'Posts': 'Title',
    'replied_to_tweet': 'Body'
})

```

```python
# Step 4: Merge Datasets Vertically
# Concatenate datasets
combined_data = pd.concat([media_data, twitter_data], ignore_index=True)

```

```python
# Step 5: Create a 'Combined' Column
# Combine the 'Title' and 'Body' into a single column for analysis
combined_data['Combined'] = combined_data['Title'].fillna('') + " " + combined_data['Body'].fillna('')
```

```python
# Step 6: Load spaCy Model and VADER Sentiment Analyzer
nlp = spacy.load("en_core_web_sm")
vader = SentimentIntensityAnalyzer()

```

```python
# Function to Extract Entities
def extract_entities(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

```

```python
# Function to Perform Sentiment Analysis
def analyze_sentiment(text):
    sentiment = vader.polarity_scores(text)
    if sentiment['compound'] >= 0.05:
        return 'Positive'
    elif sentiment['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
# Function for Topic Classification (using keyword-based approach)
def classify_topic(text):
    text = text.lower()
    if "efficacy" in text or "effective" in text:
        return "Efficacy-General"
    elif "progression free survival" in text or "pfs" in text:
        return "Progression Free Survival (PFS)"
    elif "overall survival" in text or "os" in text:
        return "Overall Survival (OS)"
    elif "safety" in text and "side effects" in text:
        return "Safety-Side Effects"
    elif "safety" in text:
        return "Safety-General"
    elif "opinion" in text or "thoughts" in text:
        return "General Opinion"
    else:
        return "Others"
```

```python
# Step 7: Apply NLP Analysis
combined_data['Entities'] = combined_data['Combined'].apply(extract_entities)
combined_data['Topic'] = combined_data['Combined'].apply(classify_topic)
combined_data['Sentiment'] = combined_data['Combined'].apply(analyze_sentiment)

```

```python
# Step 8: Save Output to CSV
combined_data.to_csv("predictions_without_openai.csv", index=False)
```
