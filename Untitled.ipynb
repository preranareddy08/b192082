```python
# Convert 'Activity Date' to datetime format
df['Activity Date'] = pd.to_datetime(df['Activity Date'], errors='coerce')

# Check for missing data
print(df.isnull().sum())

# Drop rows with invalid Activity Date or fill missing values if needed
df = df.dropna(subset=['Activity Date'])

# Optionally fill missing 'Count of Activity Metrics' with zero
df['Count of Activity Metrics'] = df['Count of Activity Metrics'].fillna(0)
```

```python
# Extract Month from Activity Date
df['Month'] = df['Activity Date'].dt.to_period('M')

# Group by Communication Source, Channel, and Month
segmented_data = df.groupby(['Communication Source', 'Source Channel Name', 'Month']).agg(
    Reach=('Physician ID', 'nunique'),
    Frequency=('Physician ID', lambda x: len(x) / x.nunique()),
    Engagement=('Count of Activity Metrics', 'sum')
).reset_index()

# Display segmented metrics
segmented_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define a function to create bar plots
def create_bar_plot(data, x_col, y_col, title, xlabel, ylabel, palette="viridis"):
    # Check if the required columns are in the DataFrame
    if x_col not in data.columns or y_col not in data.columns:
        print(f"Error: Columns '{x_col}' or '{y_col}' not found in the dataset.")
        return

    plt.figure(figsize=(10, 6))
    sns.barplot(data=data, x=x_col, y=y_col, errorbar=None, palette=palette)  # Updated parameter
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Generate plots for Reach
create_bar_plot(segmented_data, x_col='Source Channel Name', y_col='Reach',
                title='Reach by Source Channel', xlabel='Source Channel', ylabel='Reach')

# Ensure the column exists before plotting
if 'Activity' in segmented_data.columns:
    create_bar_plot(segmented_data, x_col='Activity', y_col='Reach',
                    title='Reach by Activity', xlabel='Activity', ylabel='Reach')
else:
    print("Column 'Activity' not found in the dataset. Skipping plot for Reach by Activity.")

if 'Therapeutic Area' in segmented_data.columns:
    create_bar_plot(segmented_data, x_col='Therapeutic Area', y_col='Reach',
                    title='Reach by Therapeutic Area', xlabel='Therapeutic Area', ylabel='Reach')
else:
    print("Column 'Therapeutic Area' not found in the dataset. Skipping plot for Reach by Therapeutic Area.")

# Generate plots for Frequency
create_bar_plot(segmented_data, x_col='Communication Source', y_col='Frequency',
                title='Frequency by Communication Source', xlabel='Communication Source', ylabel='Frequency', palette='cubehelix')

create_bar_plot(segmented_data, x_col='Source Channel Name', y_col='Frequency',
                title='Frequency by Source Channel', xlabel='Source Channel', ylabel='Frequency', palette='inferno')

if 'Therapeutic Area' in segmented_data.columns:
    create_bar_plot(segmented_data, x_col='Therapeutic Area', y_col='Frequency',
                    title='Frequency by Therapeutic Area', xlabel='Therapeutic Area', ylabel='Frequency', palette='magma')
else:
    print("Column 'Therapeutic Area' not found in the dataset. Skipping plot for Frequency by Therapeutic Area.")

# Generate plots for Engagement
create_bar_plot(segmented_data, x_col='Communication Source', y_col='Engagement',
                title='Engagement by Communication Source', xlabel='Communication Source', ylabel='Engagement', palette='cividis')

if 'Device' in segmented_data.columns:
    create_bar_plot(segmented_data, x_col='Device', y_col='Engagement',
                    title='Engagement by Device', xlabel='Device', ylabel='Engagement', palette='spring')
else:
    print("Column 'Device' not found in the dataset. Skipping plot for Engagement by Device.")
```

```python
import matplotlib.pyplot as plt

# Define a function to create a pie chart
def create_pie_chart(data, column, title):
    # Aggregate the data by the specified column and sum the engagement values
    engagement_data = data.groupby(column)['Engagement'].sum()

    # Create a pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(engagement_data, labels=engagement_data.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2', len(engagement_data)))
    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

# Create a pie chart for Engagement by Source Channel
create_pie_chart(segmented_data, column='Source Channel Name', title='Engagement Distribution by Source Channel')
```
